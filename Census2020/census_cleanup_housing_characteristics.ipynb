```python
import pandas as pd
```

```python
df = pd.read_csv("/Users/eavallon/Desktop/FosterCapacity/Census2020/raw/ACSDP5Y2020.DP04-housingcharacteristics.csv")
```

```python
df.columns = df.columns.str.replace("!!", " ")
df = df.replace("ZCTA5 ","",regex=True)
df = df.rename(columns={"Label (Grouping)": "Zip Code"})
```

```python
df_percent = df.copy()
```

```python
for index, row in df.iterrows():
    if index == len(df) - 3:
        df.loc[index,'HOUSING OCCUPANCY Total housing units':] = df.loc[index+1,'HOUSING OCCUPANCY Total housing units':]
        df_percent.loc[index,'HOUSING OCCUPANCY Total housing units':] = df_percent.loc[index+2,'HOUSING OCCUPANCY Total housing units':]
        break
    else:
        df.loc[index,'HOUSING OCCUPANCY Total housing units':] = df.loc[index+1,'HOUSING OCCUPANCY Total housing units':]
        df_percent.loc[index,'HOUSING OCCUPANCY Total housing units':] = df_percent.loc[index+2,'HOUSING OCCUPANCY Total housing units':]
```

```python
drop_idx = [i for i in range(1, len(df), 3)]
drop_idx_perc = [i for i in range(2, len(df), 3)]
df = df.drop(drop_idx + drop_idx_perc)
df_percent = df_percent.drop(drop_idx + drop_idx_perc)
```

```python
df.to_csv('/Users/eavallon/Desktop/FosterCapacity/Census2020/clean/ACSDP5Y2020.DP04-housingcharacteristics/housing_characteristics.csv')
df_percent.to_csv('/Users/eavallon/Desktop/FosterCapacity/Census2020/clean/ACSDP5Y2020.DP04-housingcharacteristics/housing_characteristics_percent.csv')
```
