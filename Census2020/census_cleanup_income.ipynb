```python
import pandas as pd
```

```python
df = pd.read_csv("./raw/ACSST5Y2020.S1901-income.csv").transpose()
```

```python
df
```

```python
# df.columns = df.columns.str.replace("Total:!!","")
# df.columns = df.columns.str.replace(":!!"," ")
# df.columns = df.columns.str.replace(":"," ")
df = df.rename(columns={df.columns[i]: df.iloc[0][i] for i in range(len(df.columns))})
df = df.reset_index()
df = df.replace("ZCTA5 ","",regex=True)
df = df.replace("±", "", regex=True)
df = df.rename(columns={"index": "Zip Code"})
```

```python
df = df.drop(0)
```

```python
df
```

```python
zipcodes = [df['Zip Code'][i].split('!!')[0] for i in df.index]
label = [df['Zip Code'][i].split('!!')[1] for i in df.index]
grouping = [df['Zip Code'][i].split('!!')[2] for i in df.index]
```

```python
df['Zip Code'] = zipcodes
df.insert(1,'Label',label)
df.insert(2,'Grouping',grouping)
```

```python
df
```

```python
df_err = df.copy()
```

```python
for index, row in df.iterrows():
    if index == len(df) - 3:
            df.iloc[index][1:] = df.iloc[index + 1][1:]
            df_err.iloc[index][1:] = df_err.iloc[index + 2][1:]
            break
    else:
            df.iloc[index][1:] = df.iloc[index + 1][1:]
            df_err.iloc[index][1:] = df_err.iloc[index + 2][1:]
```

```python
drop_idx = [i for i in range(2,len(df),2)]
drop_idx_err = [i for i in range(1,len(df),2)]
df = df.drop(drop_idx)
df_err = df_err.drop(drop_idx_err)
```

```python
df_err
```

```python
df.to_csv("./clean/ACSST5Y2020.S1901-income_clean.csv")
df_err.to_csv("./clean/ACSST5Y2020.S1901-income_err_clean.csv")
```
